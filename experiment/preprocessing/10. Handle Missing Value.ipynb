```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from pandas.plotting import scatter_matrix
from sklearn.impute import SimpleImputer
```

```python
housing = pd.read_csv("files/housing.csv")
```

```python
housing.info()
```

```python
housing
```

```python
# housing.drop("total_bedrooms", axis=1)
housing.dropna(subset= ["total_bedrooms"], inplace = False).info()
```

```python
median = housing["total_bedrooms"].median()
median
```

```python
housing["total_bedrooms"].fillna(median, inplace= True)
```

```python
housing.info()
```

```python
housing_num = housing.select_dtypes(include= [np.number])
```

```python
housing_num
```

```python
impute = SimpleImputer(strategy= "median")
```

```python
impute.fit(housing_num)
```

```python
impute.statistics_
```

```python
arr = impute.transform(housing_num)
```

```python
clm = impute.get_feature_names_out()
```

```python
df = pd.DataFrame(arr, columns= clm)
df
```
